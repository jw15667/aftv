```python
driver = webdriver.Chrome(PATH)
driver.maximize_window()

wait = WebDriverWait(driver, 15)
presence = EC.presence_of_element_located
visible = EC.visibility_of_element_located
data_frame = {}

for link in hyperlinks[3:4]:
    driver.get(link)
    driver.find_element_by_class_name('VfPpkd-LgbsSe VfPpkd-LgbsSe-OWXEXe-k8QpJ VfPpkd-LgbsSe-OWXEXe-dgl2Hf nCP5yc AjY5Oe DuMIQc IIdkle').click()
    # wait
    wait.until(visible((By.ID, "info-contents")))
    textbox = driver.find_element_by_id('info-contents')
    source_code = textbox.get_attribute("innerHTML")
    data_frame[link] = source_code
```
